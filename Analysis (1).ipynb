HR Attrition Analysis

 Objective

To analyze employee attrition patterns and identify key factors influencing employee turnover using the IBM HR dataset.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
```

```python
df = pd.read_csv("C:\\Users\\Admin\\OneDrive\\Desktop\\Data sets\\WA_Fn-UseC_-HR-Employee-Attrition.csv")

df.head()
df.info()
df.describe()
```

1.Total Employees

```python
total_employees = df.shape[0]
print("Total Employees:", total_employees)
```

2.Attrition Count

```python
attrition_count = df['Attrition'].value_counts()
print(attrition_count)
```

3.Attrition Rate

```python
attrition_rate = (df['Attrition'].value_counts(normalize=True)['Yes']) * 100
print("Attrition Rate: {:.2f}%".format(attrition_rate))
```

Visualization 1 – Attrition Distribution


This chart shows the overall distribution of employees who left the company versus those who stayed.
The calculated attrition rate is 16.12%.

```python
sns.countplot(x='Attrition', data=df)
plt.title("Attrition Distribution")
plt.show()
```

Insight

The overall attrition rate is 16.12%, indicating that approximately 1 in 6 employees leave the organization. While the majority of employees stay, this turnover rate suggests room for improvement in retention strategies.

Visualization 2: Attrition by Department



```python
dept_attrition = pd.crosstab(df['Department'], df['Attrition'])
dept_attrition.plot(kind='bar')
plt.title("Attrition by Department")
plt.show()
```

Insight

The Sales department shows the highest number of attrition cases compared to other departments. This may indicate higher job pressure, performance targets, or workload challenges within Sales.

Visualization 3:Attrition by Overtime

```python
sns.countplot(x='OverTime', hue='Attrition', data=df)
plt.title("Overtime vs Attrition")
plt.show()
```

Insight

Employees who work overtime exhibit a noticeably higher attrition rate compared to those who do not. This suggests that excessive workload or work-life imbalance may be a key driver of employee turnover.

Visualization Job Satisfaction vs Attrition

This analysis checks whether employee job satisfaction level has an impact on attrition.

```python
sns.countplot(x='JobSatisfaction', hue='Attrition', data=df)
plt.title("Job Satisfaction vs Attrition")
plt.show()
```

 Insight

Employees with lower job satisfaction levels (1 and 2) exhibit comparatively higher attrition rates than those with higher satisfaction levels (3 and 4). This indicates that employee engagement and workplace satisfaction are critical factors influencing retention. Strengthening employee engagement initiatives may help reduce turnover.

Final Conclusion

This analysis examined employee attrition patterns using the IBM HR dataset to identify key factors influencing turnover.

The overall attrition rate is 16.12%, indicating that approximately 1 in 6 employees leave the organization. The analysis reveals that attrition is higher in the Sales department, among employees who work overtime, and among those with lower job satisfaction levels.

These findings suggest that workload management, employee engagement, and department-specific retention strategies are critical areas for improvement. By reducing excessive overtime, strengthening employee engagement initiatives, and focusing retention efforts within high-risk departments, the organization can potentially lower attrition rates and improve workforce stability.

This project demonstrates how data-driven insights can support strategic HR decision-making and improve employee retention outcomes.
